```python
import pandas as pd
import numpy as np
```

```python
test = pd.read_csv("test.csv")
test
```

```python
df = pd.read_csv("test_data2.csv")
```

```python
df.rename({'Дата платежа': "year"}, axis=1, inplace=True)
df
```

```python
df.drop(columns="Отнесено", inplace=True)
df
```

```python
df['id'] = np.nan
df
```

```python
for name in df['Направление расходов'].unique():
    obj = test.loc[test['Направление расходов'] == name].iloc[0]
    df.loc[df['Направление расходов'] == name, 'id'] = obj['id']
df
```

```python
df['id'] = df['id'].astype(int)
df
# 4512
```

```python
list_df = pd.read_csv("list.csv")

df = df.merge(list_df, on=['id', 'year'], how="inner")
```

```python
df.drop(columns=["Направление расходов", "object_name", "ids_row", "balance_date"], inplace=True)
df
```

```python
rashod_names = df['Наименование вида расходов'].to_list()
rashod_names_dict = {rashod_names[i]: i for i in range(len(rashod_names))}

technical_conditions = df["technical_condition"].to_list()
technical_conditions_dict = {technical_conditions[i]: i for i in range(len(technical_conditions))}

otdels = df["otdel"].to_list()
otdels_dict = {otdels[i]: i for i in range(len(otdels))}

city = df["city"].to_list()
city_dict = {city[i]: i for i in range(len(city))}

df["Наименование вида расходов"] = df["Наименование вида расходов"].map(rashod_names_dict)
df["technical_condition"] = df["technical_condition"].map(technical_conditions_dict)
df["otdel"] = df["otdel"].map(otdels_dict)
df["city"] = df["city"].map(city_dict)
df
```

```python
import tensorflow as tf

model = tf.keras.models.load_model("Model")

x = np.array(df)

y = model.predict(x)
```

```python
ans = []
for i in y:
    ans.append(i[0])
ans
```

```python
submit = pd.read_csv("test_data.csv")
submit['Отнесено'] = ans
submit.to_csv("test_data.csv", index=False)
```
