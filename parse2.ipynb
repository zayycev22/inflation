```python
import pandas as pd
```

```python
expenses = pd.read_csv("expenses.csv")
expenses
```

```python
expenses.dropna(subset=['y1'], inplace=True)
expenses
```

```python
list_df = pd.read_csv("list.csv")
list_df
```

```python
kr_ks_tp = pd.read_csv("kr_ks_tp.csv")
kr_ks_tp
```

```python
kr_ks_tp.dropna(inplace=True)
kr_ks_tp['Column5'] = kr_ks_tp['Column5'].astype(int)
kr_ks_tp
```

```python
expenses.rename({'ids_row': "id"}, inplace=True, axis=1)
expenses
```

```python
expenses = expenses.merge(list_df, on=["year", "id"], how="inner")
```

```python
# dicts technical_condition otdel city rashod_name
# drop naprvalenie object_name ids_row
expenses
```

```python
expenses = expenses.drop(columns=["naprvalenie", "naprvalenie", "object_name", "ids_row"])
expenses
```

```python
rashod_names = expenses['rashod_name'].to_list()
rashod_names_dict = {rashod_names[i]: i for i in range(len(rashod_names))}

technical_conditions = expenses["technical_condition"].to_list()
technical_conditions_dict = {technical_conditions[i]: i for i in range(len(technical_conditions))}

otdels = expenses["otdel"].to_list()
otdels_dict = {otdels[i]: i for i in range(len(otdels))}

city = expenses["city"].to_list()
city_dict = {city[i]: i for i in range(len(city))}

expenses["rashod_name"] = expenses["rashod_name"].map(rashod_names_dict)
expenses["technical_condition"] = expenses["technical_condition"].map(technical_conditions_dict)
expenses["otdel"] = expenses["otdel"].map(otdels_dict)
expenses["city"] = expenses["city"].map(city_dict)
expenses
```

```python
expenses.to_csv("final.csv", index=False)
```
